## 3. DataFrame

A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or a SQL table, or a dictionary of Series objects. It is the most commonly used Pandas object.

It is a table with rows and columns, where each column has a name and a type. It is very similar to a dictionary of lists, where each list is a column of the table.

## 4. Main methods

- **pd.read_csv()** - Reads a .csv file into a DataFrame.

- **df.head()** - Returns the first 5 rows of the DataFrame.

- **df.tail()** - Returns the last 5 rows of the DataFrame.

- **df.info()** - Returns a summary of the DataFrame.

- **df.describe()** - Returns a summary of the numerical columns of the DataFrame.

- **df.shape** - Returns the number of rows and columns of the DataFrame.

- **df.columns** - Returns the names of the columns of the DataFrame.

- **df.index** - Returns the index of the DataFrame.

- **df.values** - Returns the values of the DataFrame as a numpy array.

- **df.T** - Transposes the DataFrame.

- **df.drop()** - Drops a column or row from the DataFrame.

- **df.isnull()** - Returns a DataFrame with True or False values depending on whether each value is null or not.

- **df.fillna()** - Fills the null values of the DataFrame with a given value.

- **df.loc[]** - Returns a row or rows of the DataFrame by index.

- **df.iloc[]** - Returns a row or rows of the DataFrame by position.

- **df.to_csv()** - Writes a DataFrame into a .csv file.

- **pd.read_excel()** - Reads an Excel file into a DataFrame.

- **df.to_excel()** - Writes a DataFrame into an Excel file.

- **Adding new columns** - We can add new columns to a DataFrame by assigning a list or a Series to a new column name.

- **df.groupby()** - Groups the DataFrame by a given column.

- **df.sort_values()** - Sorts the DataFrame by a given column.

- **df.apply()** - Applies a function to a column or row of the DataFrame.

- **df.unique()** - Returns the unique values of a column of the DataFrame.

#### Ex 3.

Create a new column in the previous DataFrame called "Continent" based on the city of each user. You can start by checking the unique countries and building a list of the countries for each continent. Then you can use the apply() method to apply a function to each row of the DataFrame to get the continent of each user.

```python
# Your code here

import pandas as pd

# 1. Recreate the DataFrame (since context might be fresh)
users = {
    "id": list(range(1, 21)),
    "name": ["John", "Jane", "Alice", "Bob", "Eve", "Jack", "Jill", "Kate", "Mark", "Paul", "Bob", "Alice", "Jane", "John", "John", "Alice", "Jane", "John", "Alice", "Jane"],
    "telephone": ["123456789", "987654321", "123123123", "456456456", "789789789", "321321321", "654654654", "987987987", "159159159", "753753753", "637838399", "123456789", "987654321", "123123123", "456456456", "789789789", "321321321", "654654654", "987987987", "159159159"],
    "has_car": [True, True, False, True, False, True, False, True, False, True, False, False, False, True, True, False, False, False, True, True],
    "birth_date": ["1980-01-01", "1981-02-02", "1982-03-03", "1983-04-04", "1984-05-05", "1985-06-06", "1986-07-07", "1987-08-08", "1988-09-09", "1989-10-10", "2001-01-01", "2002-02-02", "2003-03-03", "2004-04-04", "2005-05-05", "2006-06-06", "2007-07-07", "2008-08-08", "2009-09-09", "2010-10-10"],
    "city": ["Barcelona", "Paris", "London", "Madrid", "Rome", "Pekin", "Amsterdam", "Lima", "Lisbon", "Kigali", "Barcelona", "Paris", "London", "Madrid", "Rome", "Pekin", "Amsterdam", "Lima", "Lisbon", "Kigali"],
}
users_df = pd.DataFrame(users)

# 2. Check unique cities to make sure we cover all cases
unique_cities = users_df["city"].unique()
print("Unique Cities:", unique_cities)

# 3. Define the function to map city to continent
def get_continent(city):
    if city in ["Barcelona", "Paris", "London", "Madrid", "Rome", "Amsterdam", "Lisbon"]:
        return "Europe"
    elif city == "Pekin":
        return "Asia"
    elif city == "Lima":
        return "South America"
    elif city == "Kigali":
        return "Africa"
    else:
        return "Unknown"

# 4. Apply the function to create the new column
users_df["Continent"] = users_df["city"].apply(get_continent)

# 5. Show the result
print("\nUpdated DataFrame with Continent:")
print(users_df[["name", "city", "Continent"]].head(10))
```
