### Ex 1:
Write a Python program that takes a string input from the user and counts the number of vowels and consonants in the string. Display the counts to the user.


````python
# Your code here
# Get input from the user
text = input("Enter a string: ")

# Initialize counters
vowels_count = 0
consonants_count = 0

# Define vowels
vowels = "aeiouAEIOU"

# Iterate over each character in the string
for char in text:
    if char.isalpha():  # Check if the character is a letter
        if char in vowels:
            vowels_count += 1
        else:
            consonants_count += 1

# Display the results
print("Number of vowels:", vowels_count)
print("Number of consonants:", consonants_count)

````

Explanation:

[Your explanation here]

The idea is to examine each character in the user’s input, determine if it’s an alphabetical letter, classify it as either a vowel or a consonant, and keep a running count of each so we can report the totals at the end.

Write a Python function that takes a list of integers as input and returns a new list containing only the numbers which are multiples of 3 from the original list. Use and not use list comprehension to achieve this.


````python
numbers = [32, 19, 45, 60, 2, 880, 91, 141, 0, 73, 54]

# Your code here
# Function to filter multiples of 3
# Use list comprehension
# def multiples_of_three(numbers):
#     return [num for num in numbers if num % 3 == 0]

# # Example list
# numbers = [32, 19, 45, 60, 2, 880, 91, 141, 0, 73, 54]

# # Call the function and print the result
# result = multiples_of_three(numbers)
# print(result)

# Without using list comprehension
def multiples_of_three(numbers):
    result = []
    for num in numbers:
        if num % 3 == 0:
            result.append(num)
    return result

# Example list
numbers = [32, 19, 45, 60, 2, 880, 91, 141, 0, 73, 54]

# Call the function and print the result
print(multiples_of_three(numbers))
````

Explanation:

[Your explanation here]

The idea is simple — we look at each number in the list, ask “is this number divisible by 3?” (that’s what num % 3 == 0 does), and if the answer is yes, we keep it. Then we put all those numbers together in a new list. So basically, we’re just picking out all the numbers that are multiples of 3.

### Ex 4:
Write a Python program that takes as input the following text fragment and counts the frequency of each different word in the file. Display the top 5 most frequent words along with their using a dictionary to store the word counts. 

```python
# For example:
{"hello": 2, "world": 67, ...}
```


````python

# Your code here

# Given text
text = """In computer science, a data structure is a data organization, storage, and retrieval format 
that enables efficient access and modification of data. Common data structures include arrays, linked 
lists, stacks, queues, trees, and graphs. Each data structure has its own strengths and weaknesses, 
making them suitable for different types of applications and algorithms. Choosing the right data structure 
is crucial for optimizing performance and resource utilization in software development."""

# Step 1: Clean and split the text into words
# Convert to lowercase so "Data" and "data" count as the same word
# Remove punctuation for clean counting
import string

clean_text = text.lower()
for punct in string.punctuation:
    clean_text = clean_text.replace(punct, "")

words = clean_text.split()

# Step 2: Count word frequencies using a dictionary
word_counts = {}

for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Step 3: Sort by frequency (highest first)
sorted_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Step 4: Display the top 5 most frequent words
print("Top 5 most frequent words:")
for word, count in sorted_counts[:5]:
    print(f"{word}: {count}")

# Example dictionary output (full version)
#print(word_counts)
````

Explanation:

[Your explanation here]

We clean up the text — turn everything lowercase and remove punctuation so we don’t count “data” and “data,” as different words.
Then we split the text into a list where each item is a word.
We loop through these words, and for each one, we either start its count or increase it in a dictionary.
The dictionary ends up looking something like:
{"data": 5, "structure": 4, "and": 6, ...}
Finally, we sort that dictionary by how many times each word appears and print the top 5.

## Ex 5:

Write a Python program that manages a simple student grade system. Create a function that takes a list of tuples as input, where each tuple contains a student's name and their score (e.g., `("Alice", 85)`). The function should:

1. Calculate and return the average score of all students
2. Find and return the student with the highest score
3. Return a list of tuples containing only students who scored above the average

Use the provided sample data to test your function. Display all three results clearly.


````python
students = [
    ("Alice", 85),
    ("Bob", 72),
    ("Charlie", 91),
    ("Diana", 78),
    ("Eve", 88),
    ("Frank", 65),
    ("Grace", 95)
]

# Your code here

# Function to process student grades
def process_grades(students):
    # 1. Calculate average score
    total = 0
    for name, score in students:
        total += score
    average = total / len(students)

    # 2. Find student with the highest score
    highest_student = students[0]
    for student in students:
        if student[1] > highest_student[1]:
            highest_student = student

    # 3. Get list of students who scored above average
    above_average = []
    for student in students:
        if student[1] > average:
            above_average.append(student)

    return average, highest_student, above_average


# Sample data
students = [
    ("Alice", 85),
    ("Bob", 72),
    ("Charlie", 91),
    ("Diana", 78),
    ("Eve", 88),
    ("Frank", 65),
    ("Grace", 95)
]

# Run the function
avg, top_student, above_avg = process_grades(students)

# Display results
print(f"Average Score: {avg:.2f}")
print(f"Highest Scoring Student: {top_student[0]} with {top_student[1]}")
print("Students scoring above average:")

for name, score in above_avg:
    print(f"{name}: {score}")
````

Explanation:

[Your explanation here]

First, we go through all the students and add up their scores, then divide by how many students there are to get the average.
Next, we look for the student with the highest score by comparing each student to the current best one.
After that, we check who scored higher than the average and put them into a list.
Finally, we print everything in a nice, readable way.
